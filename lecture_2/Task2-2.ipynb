```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy
from scipy import linalg 
```

```python
def f(x):
    return math.sin(x / 5.0) * math.exp(x / 10.0) + 5.0 * math.exp(-x / 2.0)

def fp(x, params):
    i = 0
    result = 0
    for param in params:
        result = result + param * (x ** i)
        i = i + 1
    return result

def rngParam(val):
    rank = len(values)-1
    return [val ** n for n in range(0, rank + 1)]

def write(str, name):
    file_obj = open(name, 'w')
    file_obj.write(str)
    file_obj.close()
    
# 2
values = [1, 15]
M = [rngParam(value) for value in values]
V = [f(x) for x in values]
S = scipy.linalg.solve(M, V)

V = [fp(x, S) for x in values]
plt.plot(values, V)

# 3
values = [1, 8, 15]
M = [rngParam(value) for value in values]
V = [f(x) for x in values]
S = scipy.linalg.solve(M, V)

V = [fp(x, S) for x in values]
plt.plot(values, V)

# 4
values = np.arange(1.,15.,2)
values = [1, 4, 10, 15]
M = [rngParam(value) for value in values]
V = [f(x) for x in values]
S = scipy.linalg.solve(M, V)

V = [fp(x, S) for x in values]
plt.plot(values, V)
#write(' '.join([str(p) for p in S]))

Xs = np.arange(1.,15.,1)
V = [f(x) for x in Xs]

plt.plot(Xs, V)
plt.grid(True)
plt.show()
```
